```python
#Q1 find all unique vowel present in a string using a set.
text = "programimg in python "
#output=---->{'o','i',a}
b=set()
for i in text:
  if i in 'aeiou':
    b.add(i)
print(b)
```

```python
#Q2 given 2 list ,find common elements using set.
list1=[1,2,3,4,5,6]
list2=[4,5,6,7,8,9]
#output--->{4,5,6}
a=set(list1)
b=set(list2)
c=a&b
print(c)
```

```python
#Q3 find elemts that are in list1 but not in list2.
l1=[1,2,3,4]
l2=[3,4,5,6]
#output--->{1,2}
a=set(l1)
b=set(l2)
print(a-b)
```

```python
#Q4 find all duplicate elements in a list using a set .
# nums=[1,2,3,2,4,5,1,6,3]
#output--->{1,2,3}
nums = [1, 2, 3,2, 4, 5, 1, 6, 3]
a = set()
b = set()

for num in nums:
    if num in a:
        b.add(num)
    else:
        a.add(num)

print(b)
```

```python
#Q5 check if ll elements of one list exits in another list using sets.
l1=[1,2,3]
l2=[1,2,3,4,5]
#output---> true
l1=set(l1)
l2=set(l2)
if l1.issubset(l2):
    print("true")
else:
    print("false")
```
