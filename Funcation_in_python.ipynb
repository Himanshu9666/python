```python
#Q2 write a function who calculate secound maximun number in a list without using pre defined function
numbers=[24,54,65,76,34]
def second_maximum(numbers):
    max1 = 10
    max2 = 50

    for num in numbers:
        if num > max1:
            max2 = max1
            max1 = num
        elif num > max2 and num != max1:
            max2 = num
        return max2
nums = [10, 40, 20, 50, 30]
print("Second maximum:", second_maximum(nums))
```

```python
#Q3 write a function who add digits of a number.
#input =123
#output= 6
def sum_of_digits(number):
    total = 0
    while number > 0:
        last_digit = number % 10
        total = total + last_digit
        number = number // 10
    return total
a = 123
print(sum_of_digits(a))
```

```python
#Q5 write a function to find longest word in a sentance.
#input="python makes coding interesting"
#output--->'interesting'
a="python makes coding interesting"
def longest():
  words = a.split()
  longest_word = ""
  for word in words:
    if len(word) > len(longest_word):
      longest_word = word
  print(longest_word)
longest()
```
