```python
#first and secund mx number
numbers = [12, 45, 67, 23, 89, 34]

first_max = second_max = float('-inf')

for num in numbers:
    if num > first_max:
        second_max = first_max
        first_max = num
    elif num > second_max and num != first_max:
        second_max = num

print("First maximum:", first_max)
print("Second maximum:", second_max)
```

```python
#calculate the sum of first 15 odd numbers using while loop?

count = 0
number = 1
sum_of_odds = 0

# Loop until we find 15 odd numbers
while count < 15:
    if number % 2 != 0:    # Check if the number is odd
        sum_of_odds += number
        count += 1
    number += 1            # Move to the next number

# Print the result
print("Sum of first 15 odd numbers is:", sum_of_odds)

```

```python
# Step 1: Create user-defined list
a = []
size = int(input("Enter size of the list: "))
for i in range(size):
    val = int(input("Enter number: "))
    a.append(val)

print("Original List:", a)

# Step 2: Create list with unique elements
b = []
for i in a:
    if i not in b:
        b.append(i)

print("Unique List:", b)

# Step 3: Find maximum and second maximum
if len(b) >= 2:
    max1 = max(b)
    b_copy = b.copy()
    b_copy.remove(max1)
    max2 = max(b_copy)
    print("Maximum number:", max1)
    print("Second Maximum number:", max2)
elif len(b) == 1:
    print("Only one unique number. Maximum is:", b[0])
    print("Second Maximum does not exist.")
else:
    print("No elements in the list.")

# Step 4: Find total sum of unique list
total_sum = sum(b)
print("Total sum of unique elements:", total_sum)
```

```python
# Step 1: Create a user-defined list
a = []
size = int(input("Enter size of the list: "))
for i in range(size):
    val = input("Enter element: ")  # Use input() to allow both numbers and letters
    a.append(val)

print("Original List:", a)

# Step 2: Reverse the list without using reverse() or slicing
reversed_list = []
for i in range(len(a)-1, -1, -1):
    reversed_list.append(a[i])

print("Reversed List:", reversed_list)

# Step 3: Check if the list is a palindrome
if a == reversed_list:
    print("The list is a palindrome.")
else:
    print("The list is not a palindrome.")
```

```python
#Q1 shop --> [pen,notebook,pizza,icecream,vdapav,football,jeans,shoes,shirt]
#filter out -->stationry -->[], bakery-->,cloths-->[]

shop = ['pen','notebook','pizza','icecream','vdapav','football','jeans','shoes','shirt']
stationry=[]
bakery=[]
cloths=[]

for i in shop:
  if i in ['pen','notebook','football']:
    stationry.append(i)
  elif i in ['pizza','icecream','vdapav']:
    bakery.append(i)
  else:
    cloths.append(i)

print("Stationery:", stationry)
print("Bakery:", bakery)
print("Cloths:", cloths)
```

```python
#Q5 clean list by removing special characters
# a=['py@thon','d#ata','sci!ence']

#result-->['python','data','science']
a = ['py@thon', 'd#ata', 'sci!ence']

cleaned = []
for word in a:
    new_word = ""
    for ch in word:
        if ch.isalpha():   # keep only letters
            new_word += ch
    cleaned.append(new_word)

print(cleaned)
```
