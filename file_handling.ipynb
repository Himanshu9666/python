```python
# file not found error
# try block also run always...
try:
  file = open('sample.txt','r')
  content = file.read()
# it only works when try block gives an error
except FileNotFoundError:
  print('The file do not exist')
# it will only run when try block do not gives error
else:
  print(content)
  file.close()
# this block always run
finally:
  print('This block will always run')
```

```python
from logging import exception
# BANK MANAGEment
Balance = 1000000

def debit(amount):
  global Balance
  if amount > Balance:
    raise exception('Insufficient Balance')
  else:
    Balance = Balance - amount
    print(f'${amount} is debited')
    print(f'${Balance}')
try:
  debit(234432)
except exception as e:
  print('Insufficient Balance')
```
