```python
# A decorator is essentially a function that takes another function as an argument
# and returns a new function with enhanced functionality

#Decoraters

def display():
  print("this is the display funcation !")
display()
```

```python
def my_decorator(display):
  def wrapper():
    print("this task run before the main funcation runs")
    display()
    print("this is the task which will run after the main funcation")
  return wrapper

@my_decorator
def display():
  print("thi is another task")
  print("this is the main funcation")

display()
```

```python
def decorator(func):
  def wrapper(a,b):
    print("this task will run before the main task runs")

    func(a,b)
    print("this is the task which will run after the main task")
  return wrapper

@decorator
def multiply(a,b):
  print(a*b)

multiply(2,3)
```

```python
def decorator(func):
  def wrapper(a,b):
    print("this task will run before the main task runs")
    a=a+b
    func(a,b)
    print("this is the task which will run after the main task")
  return wrapper

@decorator
def multiply(a,b):
  print(a*b)

def add(a,b):
  print(a+b)

multiply(2,3)
add(5,6)
```

```python
# iterators
# In Python, an iterator is an object that enables sequential traversal through the elements of a collection,
# such as lists, tuples, dictionaries, and sets. Iterators implement the iterator protocol, which consists of two special methods
li = [x for x in range(1000)]
iterable = iter(li)

print(next(iterable))
print(next(iterable))
print(next(iterable))
print(next(iterable))
print(next(iterable))
```

```python
#genrators
#genrators = functions that remeber where they left off and produce one value at a time
#yield = pause+rember+return value
#genrater = lightweight iterators
#used when you want to procces data laily (one-demand) inseted of all at one

def count(n):
  for i in range(1,n+1):
    yield i
gen = count(3)
# if we are perfoming print task it will give error  StopIteration
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
#Genrator Questions
# Write a generator function named count_up_to_five that yields numbers from 1
# to 5. Use next() to print each number produced by this generator.

def count_up_to_five():
    for i in range(1, 6):
        yield i


numbers = count_up_to_five()

print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))


```

```python
# Q3. Implement a generator function named even_numbers_up_to_ten that yields
# even numbers up to 10. Print the first three values yielded by this generator using
# next().
def even_numbers_up_to_ten():
    for i in range(1,11):
      if i%2==0:
        yield i


even_numbers = even_numbers_up_to_ten()

print(next(even_numbers))
print(next(even_numbers))
print(next(even_numbers))
```

```python
# Q4. Design a generator function called string_characters that takes a string (e.g.,
# "PYTHON") and yields its characters one by one. Use next() to print the first
# three letters.
def string_characters(n):
    for i in n:
        yield i


characters = string_characters("PYTHON")

print(next(characters))
print(next(characters))
print(next(characters))
```

```python
#Q6 Write a generator function named reverse_countdown that yields numbers in
# reverse order, from 5 down to 1.
def reverse_countdown():
    for i in range(5, 0, -1):
        yield i


countdown = reverse_countdown()

for num in countdown:
    print(num)
```

```python
#Q7. Construct a generator function called vowels_only that takes a string (e.g.,
# "education") and yields only the vowels from it. Print all the yielded vowels using a
# for loop.
def vowels_only(n):
    vowels = "aeiouAEIOU"
    for i in n:
        if i in vowels:
            yield i


vowels = vowels_only("education")

for vowel in vowels:
    print(vowel)
```
